```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics  # for Silhouette Coefficient
```

```python
data = pd.read_csv('D:\Dataset\csvDataset\iris.csv')
print(data.shape)
data.head()
```

```python
data=data.loc[:,["Sepal Length","Sepal Width","Petal Length","Petal Width"]]
print(data.shape)
data.head()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans = kmeans.fit(data)
print("KMeans Clusters: ",kmeans)
labels=kmeans.labels_
labels
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
#marker size	s=	ax.scatter(x, y, s=10)
#marker color	c=	ax.scatter(x, y, c=(122, 80, 4))
#marker opacity	alpha=	ax.scatter(x, y, alpha=0.2)

plt.scatter(data["Sepal Length"], data["Sepal Width"],  s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(data, labels))
```

```python
from sklearn.datasets import make_moons
#Generate 2d classification dataset
#noise  double or None (default=None).......Standard deviation of Gaussian noise added to the data.

moon_X, moon_y = make_moons(n_samples=2000, noise=0.05) 
print("Moon sample data: ",moon_X)
print("Cluster Labels: ",moon_y)
```

```python
# scatter plot, dots colored by class value
df = pd.DataFrame(dict(x=moon_X[:,0], y=moon_X[:,1], label=moon_y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
def add_noise(X,y, noise_level = 0.04):
    #The number of points we wish to make noisy
    amt_noise = int(noise_level*len(y))
    #Pick amt_noise points at random
    idx = np.random.choice(len(X), size = amt_noise)
    #Add random noise to these selected points
    noise = np.random.random((amt_noise, 2) ) -0.5
    X[idx,:] += noise
    return X 

moon_X=add_noise(moon_X,moon_y)
```

```python
# scatter plot, dots colored by class value
df = pd.DataFrame(dict(x=moon_X[:,0], y=moon_X[:,1], label=moon_y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(moon_X)
labels=kmeans.labels_
labels
```

```python
kmeansPredicted = kmeans.predict(moon_X)
kmeansPredicted
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.scatter(moon_X[:, 0], moon_X[:, 1], c=kmeansPredicted, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(moon_X, labels))
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(moon_X)
labels=kmeans.labels_
print("Kmeans Cluster Labels: ", labels)

# predict the cluster
kmeansPredicted = kmeans.predict(moon_X)

# center of the clusters
centers = kmeans.cluster_centers_
print("center of the clusters: ", centers)

# plot the cluster sdata samples
plt.scatter(moon_X[:, 0], moon_X[:, 1], c=kmeansPredicted, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

# Silhouette Coefficient
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(moon_X, labels))
```

```python
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
#X,y
```

```python
# scatter plot, dots colored by class value
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
labels=kmeans.labels_
#print("Kmeans Cluster Labels: ", labels)

# predict the cluster
kmeansPredicted = kmeans.predict(X)

# center of the clusters
centers = kmeans.cluster_centers_
#print("center of the clusters: ", centers)

# plot the cluster sdata samples
#cmap=['viridis', 'plasma', 'inferno', 'magma', 'cividis']

plt.scatter(X[:, 0], X[:, 1], c=kmeansPredicted, s=50, cmap='plasma')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

#Silhouette Coefficient
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(X, labels))
```

```python
kmeans = KMeans(n_clusters=4)
kmeans=kmeans.fit(X)
# predict the cluster
y_pred= kmeans.predict(X)

#cluster labels
labels = kmeans.labels_

plt.scatter(X[:,0], X[:,1], c=y_pred)

#Silhouette Coefficient
print("Silhouette Coefficient: %0.3f" % metrics.silhouette_score(X, labels))
```
